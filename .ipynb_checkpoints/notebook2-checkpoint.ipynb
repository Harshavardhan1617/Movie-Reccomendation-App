```python
import pandas as pd
import sqlite3 
```

```python
conn = sqlite3.connect('E:\data\imdb\movies.sqlite')
```

```python
query = """
SELECT tconst AS movie_id
,primaryTitle AS title
,genres 
FROM repo
"""
```

```python
df = pd.read_sql_query(query, conn)
```

```python
df
```

```python
query2 = """
SELECT tweet,
conversation_id
FROM tweets
"""
```

```python
df2 = pd.read_sql_query(query2, conn)
conn.close()
df2
```

```python
df2 = df2.dropna()
```

```python
df2 = df2[df2['tweet'].str.contains('rated')]
```

```python
df2
```

```python
df2[['title','year','rating']] = df2['tweet'].str.extract(r'rated (.*) \((.*)\) (.*)')
df2
```

```python
df_1 = df2[df2['year'].str.len()>4]
df_1
```

```python
df2 = df2[~df2['conversation_id'].isin(df_1['conversation_id'])]
df2
```

```python
mask = df['title'].isin(df2['title'])
identical_count = mask.sum()
```

```python
identical_count
```

```python
movies_df = pd.merge(df[['movie_id','title','genres']], df2[['title','year','rating']], on='title')
```
