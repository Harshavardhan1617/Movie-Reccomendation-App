```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("movies.sqlite")
cursor = conn.cursor()
```

```python
query = """
SELECT *
from movies
"""
```

```python
df = pd.read_sql_query(query, conn)
```

```python
df
```

```python
df = df.drop(columns = 'index')
```

```python

df
```

```python
df.to_sql("movies", conn, if_exists="replace", index = False)

```

```python
conn.close()
```
